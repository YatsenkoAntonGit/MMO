```python
import pandas as pd
from math import sqrt
from sklearn import linear_model
```

```python
df  = pd.read_csv('Desktop/avocado.csv')
```

```python
df
```

```python
df.plot.scatter(x='AveragePrice',
                      y='Total Volume',
                      c='DarkBlue')
```

```python
df['Total  Volume sqrtNormalized'] = df['Total Volume'].map(lambda x: sqrt(x))
```

```python
df
```

```python
clf = linear_model.Lasso(alpha=0.1)
```

```python
clf.fit(df[['Total  Volume sqrtNormalized', 'Total Bags', 'year']], df['AveragePrice'])
```

```python
print(clf.coef_)
```

```python
{key:value for key,value in list(zip(['Total  Volume sqrtNormalized', 'Total Bags', 'year'], clf.coef_))}
```
